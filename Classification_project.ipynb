```
# Required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (assuming it's already loaded into 'dataset')
# dataset = pd.read_csv('Loan.csv')

# Identify continuous and categorical variables
continuous_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
categorical_columns = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 'Property_Area', 'Loan_Status']

# Summary statistics of continuous variables
print("Summary statistics of continuous variables:")
print(dataset[continuous_columns].describe())

# Categorical variable analysis
print("\nFrequency of categorical variables:")
for col in categorical_columns:
    print(f"\n{col}:")
    print(dataset[col].value_counts())

# Distribution plots for continuous variables
plt.figure(figsize=(15, 5))
for i, col in enumerate(continuous_columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(dataset[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Boxplots for continuous variables (to detect outliers)
plt.figure(figsize=(15, 5))
for i, col in enumerate(continuous_columns):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x=dataset[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# Countplot for categorical variables
plt.figure(figsize=(20, 10))
for i, col in enumerate(categorical_columns):
    plt.subplot(3, 3, i+1)
    sns.countplot(x=col, data=dataset)
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Correlation heatmap (for continuous variables)
plt.figure(figsize=(10, 6))
corr_matrix = dataset[continuous_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Continuous Variables')
plt.show()

# Pairplot for continuous variables
sns.pairplot(dataset[continuous_columns], diag_kind='kde', corner=True, plot_kws={'s': 15})
plt.suptitle('Pairplot of Continuous Variables', y=1.02)
plt.show()


```

# Outlier Removal

```
continuous_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
def remove_outliers(df, column):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for each continuous column
for col in continuous_columns:
    dataset = remove_outliers(dataset, col)
```

```
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score

# Initialize lists to store results
model_names = []
test_accuracies = []
mean_cross_val_scores = []
std_cross_val_scores = []

# Function to perform cross-validation and calculate mean and standard deviation
def evaluate_model(model, model_name, X_train, y_train, X_test, y_test):
    # Fit model
    model.fit(X_train, y_train)

    # Predict on the test set
    test_accuracy = model.score(X_test, y_test)

    # Perform cross-validation (use 10-fold CV in this example)
    cv_scores = cross_val_score(model, X_train, y_train, cv=10)

    # Append the results
    model_names.append(model_name)
    test_accuracies.append(test_accuracy)
    mean_cross_val_scores.append(np.mean(cv_scores))
    std_cross_val_scores.append(np.std(cv_scores))

# Example models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

# Assuming X_train, X_test, y_train, y_test have already been defined

# Evaluate Logistic Regression
evaluate_model(LogisticRegression(random_state=0), "Logistic Regression", X_train, y_train, X_test, y_test)

# Evaluate K-Nearest Neighbors
evaluate_model(KNeighborsClassifier(n_neighbors=3), "K-Nearest Neighbors", X_train, y_train, X_test, y_test)

# Evaluate Support Vector Classifier
evaluate_model(SVC(kernel='linear', random_state=0), "Support Vector Machine", X_train, y_train, X_test, y_test)

# Evaluate Kernel SVM
evaluate_model(SVC(kernel='rbf', random_state=0), "Kernel SVM", X_train, y_train, X_test, y_test)

# Evaluate Decision Tree
evaluate_model(DecisionTreeClassifier(random_state=0), "Decision Tree", X_train, y_train, X_test, y_test)

# Evaluate Random Forest
evaluate_model(RandomForestClassifier(random_state=0), "Random Forest", X_train, y_train, X_test, y_test)

# Evaluate Naive Bayes
evaluate_model(GaussianNB(), "Naive Bayes", X_train, y_train, X_test, y_test)

# Create a DataFrame with shorter column headers
results_df = pd.DataFrame({
    'Model': model_names,
    'Test Acc (%)': [acc * 100 for acc in test_accuracies],
    'CV Mean (%)': [mean * 100 for mean in mean_cross_val_scores],
    'CV Std (%)': [std * 100 for std in std_cross_val_scores]
})

# Round to two decimal places and display the table
results_df = results_df.round(2)
print(results_df)
```
